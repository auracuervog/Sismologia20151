## Hipocentro evento 2

```python
tp = 61.63
ts = 69.9
#Distancia hipocentral --> hipotenusa del triangulo
dh = 8*(ts-tp)
#Distancia horizontal de la estación al centro del epicentro. Calculada con la regla de Google earth
de = 14.46 #km
#Usando pitagoras:
hi = sqrt(pow(dh,2)-pow(de,2))
print hi
```
